# Student Performance — Exploratory Data Analysis

This notebook performs basic EDA on `student_performance_100k.csv`: loading the data, showing summary statistics, and creating a few visualizations (histogram, boxplot, and correlation heatmap).

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="whitegrid", rc={"figure.dpi": 100})
```

```python
# Load dataset (adjust path if needed)
df = pd.read_csv('student_performance_100k.csv')
df.shape
```

```python
# Quick preview and info
display(df.head())
print('
Info:')
df.info()
```

```python
# Descriptive statistics for numeric fields
df.describe().T
```

```python
# Histogram of final scores
plt.figure(figsize=(8,4))
sns.histplot(df['final_score'], bins=40, kde=True, color='C0')
plt.title('Distribution of Final Scores')
plt.xlabel('Final Score')
plt.ylabel('Count')
plt.show()
```

```python
# Boxplot: study time by final result
plt.figure(figsize=(6,4))
sns.boxplot(x='final_result', y='study_time_hours_per_week', data=df, palette='Set2')
plt.title('Study Time by Final Result')
plt.xlabel('Final Result')
plt.ylabel('Study Time (hours/week)')
plt.show()
```

```python
# Correlation heatmap for numeric features
num_cols = ['age','family_income','study_time_hours_per_week','absences','health','alcohol_weekend','alcohol_weekday','free_time','failures','past_grade_avg','final_score']
corr = df[num_cols].corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Matrix (numeric features)')
plt.show()
```
